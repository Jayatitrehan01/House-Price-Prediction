```python

cat_columns = ['MSSubClass', 'MSZoning', 'Street', 'LotShape', 'LandContour', 'LotConfig',
               'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle',
               'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'ExterQual',
               'ExterCond', 'Foundation', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1',
               'BsmtFinType2', 'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
               'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',
               'PavedDrive', 'SaleType', 'SaleCondition']

house_price[cat_columns] = house_price[cat_columns].astype('category')


house_price['GarageYrBlt'] = house_price['GarageYrBlt'].astype('Int64')
```

```python
house_price['LotFrontage'] = house_price['LotFrontage'].fillna(house_price['LotFrontage'].median())
house_price['MasVnrArea'] = house_price['MasVnrArea'].fillna(0)
house_price['GarageYrBlt'] = house_price['GarageYrBlt'].fillna(house_price['YearBuilt'])

# Fill categorical features
cat_cols = ['MasVnrType', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 
            'BsmtFinType1', 'BsmtFinType2', 'FireplaceQu', 
            'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']

for col in cat_cols:
    if house_price[col].dtype.name == 'category':
        if 'None' not in house_price[col].cat.categories:
            house_price[col] = house_price[col].cat.add_categories('None')
    house_price[col] = house_price[col].fillna('None')

```

DROPPING COLUMNS

```python
house_price.drop(['Electrical', 'LotFrontage'], axis=1, inplace=True)
```

CHECKING FOR NULL VALUES

```python
print(house_price.isnull().sum())
```

FEATURE ENGINEERING

```python
# Total sqft of house
house_price['TotalSF'] = house_price['TotalBsmtSF'] + house_price['1stFlrSF'] + house_price['2ndFlrSF']

# Total number of bathrooms (weighted)
house_price['TotalBath'] = (house_price['FullBath'] + house_price['HalfBath'] * 0.5 +
                            house_price['BsmtFullBath'] + house_price['BsmtHalfBath'] * 0.5)

# House age
house_price['HouseAge'] = 2025 - house_price['YearBuilt']

# Whether the house was remodeled
house_price['Remodeled'] = (house_price['YearBuilt'] != house_price['YearRemodAdd']).astype(int)

# Is the house newly built?
house_price['IsNew'] = (house_price['YearBuilt'] == house_price['GarageYrBlt']).astype(int)
```

ENCODED CATEGORICAL VARIABLES

```python
from sklearn.preprocessing import LabelEncoder

# Encode ordinal columns
ordinal_cols = ['ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond',
                'KitchenQual', 'HeatingQC', 'FireplaceQu',
                'GarageQual', 'GarageCond']
for col in ordinal_cols:
    house_price[col] = LabelEncoder().fit_transform(house_price[col])

# One-hot encode the remaining categorical variables
house_price = pd.get_dummies(house_price, drop_first=True)
```

SCALING NUMERICAL VARIABLES

```python
from sklearn.preprocessing import StandardScaler

X = house_price.drop('SalePrice', axis=1)
y = house_price['SalePrice']

numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns
scaler = StandardScaler()
X[numeric_cols] = scaler.fit_transform(X[numeric_cols])
```
